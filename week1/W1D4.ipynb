```python
import re
def clean_sentence(sentence):
    return re.sub(r'[^A-Za-z]', '', sentence.lower())
    
    
def is_palindrome(sentence):
    sentence = clean_sentence(sentence)
    if not sentence:
        return True
    first = sentence[0]
    last = sentence[-1]
    rest = sentence[1:-1]
    if first == last:
        return is_palindrome(rest)
    return False
```

```python
def reverse_sentence(sentence):
    if len(sentence) <= 1:
        return sentence
    
    first = sentence[0]
    rest = sentence[1:]

    return reverse_sentence(rest) + first
    
def is_palindrome(sentence):
    sentence = clean_sentence(sentence)
    return sentence == reverse_sentence(sentence)
```

```python
def is_palindrome(sentence):
    sentence = clean_sentence(sentence)
    while sentence:
        if sentence[0] != sentence[-1]:
            return False
        sentence = sentence[1:-1]
    return True
```

```python
def is_palindrome(sentence):
    sentence = clean_sentence(sentence)
    for index in range(len(sentence)):
        if sentence[index] != sentence[-(index+1)]:
            return False
    return True
```

```python
def is_palindrome(sentence):
    sentence = clean_sentence(sentence)
    return sentence == sentence[::-1]
```

```python
def test_even_numbers():
    assert is_palindrome('toot') == True

def test_odd_numbers():
    assert is_palindrome('tot') == True

def test_simple_values():
    assert is_palindrome('stunt nuts') == True

def test_complete_sentences():
    assert is_palindrome('Lisa Bonet ate no basil.') == True

def test_complex_sentences():
    assert is_palindrome('A man, a plan, a cat, a ham, a yak, a yam, a hat, a canal: Panama!') == True

def test_multiple_sentences():
    assert is_palindrome('Doc, note, I dissent. A fast never prevents a fatness. I diet on cod.') == True

def test_non_palindromes():
    assert is_palindrome('i am not a palindrome') == False

test_even_numbers()
test_odd_numbers()
test_simple_values()
test_complete_sentences()
test_complex_sentences()
test_multiple_sentences()
test_non_palindromes()
```

```python
def input_hell():
    user_in = input("enter something ")
    int_count = 0
    str_count = 0
    while user_in:
        try:
            int(user_in)
            int_count += 1
        except:
            str_count += 1
        user_in = input("enter something ")
    return (int_count, str_count)


```

```python
ints, strings = input_hell()
print("the user entered {} ints and {} strings".format(ints, strings))
```

```python
def build_phonebook():
    pb = {}
    
    while True:
        name = input("enter a name")
        if name == "":
            break
        number = input("enter a phone number")
        pb[name] = number
    return pb

def search_phonebook(pb):
    while True:
        who = input("Who would you like to search for? ")
        if who == "":
            break
        if who in pb:
            print(pb[who])
        else:
            print("I don't know who that is")
```

```python
book = build_phonebook()
search_phonebook(book)
```

```python
conc = {}
with open("alice_in_wonderland.txt", "r") as f:
    for line in f:
        words = line.split()
        for word in words:
            word = clean_sentence(word)
            if word in conc:
                conc[word] += 1
            else:
                conc[word] = 1
print(conc)
```

```python
#alphabetical by key
for key, value in sorted(conc.items()):
    print(key, value)
```

```python
#sorted by value
import operator
for key, value in sorted(conc.items(), key=operator.itemgetter(1), reverse=True):
    print(key, value, sep="\t")
```
